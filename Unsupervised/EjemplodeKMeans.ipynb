# Ejemplo de K-Means

Este set de datos son los viajes de Uber en NYC

```python
import pylab as plt
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
%matplotlib inline 
filename = 'data_clustering/uber-raw-data-jul14.csv'
df = pd.read_csv(filename)
```

```python
df.head()
```

```python
df = df[['Lon', 'Lat']]
```

```python
df.plot(kind='scatter', x='Lon', y='Lat');
```

```python
import folium
from folium.plugins import HeatMap
hmap = folium.Map(location=[42.5, -75.5], zoom_start=7, )
max_amount = 5
hm_wide = HeatMap( list(zip(df.Lat.values, df.Lon.values)),
                   min_opacity=0.2,
                   max_val=max_amount,
                   radius=17, blur=15, 
                   max_zoom=1, 
                 )

hmap.add_child(hm_wide)
```
